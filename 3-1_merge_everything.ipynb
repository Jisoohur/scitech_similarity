# **Bert similarity 표준화, 제곱근, 절댓값 추가**

```python
import pandas as pd
final_df = pd.read_csv('3-1_final_data.csv')
final_df
```

```python
final_df['bert_similarity_normalized'] = (final_df['bert_similarity'] + 1) / 2
```

```python
final_df['bert_similarity_abs'] = final_df['bert_similarity'].abs()
```

```python
final_df['bert_similarity_squared'] = final_df['bert_similarity'] ** 2
```

```python
final_df.to_csv('3-1_final_data.csv', index=False)
```
